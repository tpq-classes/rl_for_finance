## Tossing a Biased Coin

```python
!git clone https://github.com/tpq-classes/rl_for_finance.git
import sys
sys.path.append('rl_for_finance')
```

```python
import numpy as np
```

```python
ssp = [1, 1, 1, 1, 0]
```

```python
asp = [1, 0]
```

```python
def epoch():
    tr = 0
    for _ in range(100):
        a = np.random.choice(asp)
        s = np.random.choice(ssp)
        if a == s:
            tr += 1
    return tr
```

```python
rl = np.array([epoch() for _ in range(15)])
rl
```

```python
rl.mean()
```

```python
ssp = [1, 1, 1, 1, 0]
```

```python
def epoch():
    tr = 0
    asp = [0, 1]
    for _ in range(100):
        a = np.random.choice(asp)
        s = np.random.choice(ssp)
        if a == s:
            tr += 1
        asp.append(s)
    return tr
```

```python
rl = np.array([epoch() for _ in range(15)])
rl
```

```python
rl.mean()
```

```python
def epoch():
    tr = 0
    for _ in range(1000):
        a = np.random.choice(asp)
        s = np.random.choice(ssp)
        if a == s:
            tr += 1
    return tr
```

```python
rl = np.array([epoch() for _ in range(15)])
rl
```
