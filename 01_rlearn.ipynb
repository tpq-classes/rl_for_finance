## Supervised Learning

### Estimation

```python
def f(x):
    return 2 * x ** 2 - x ** 3 / 3
```

```python
x = np.linspace(-2, 4, 25)
x
```

```python
y = f(x)
y
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'ro');
```

```python
beta = np.cov(x, y)[0, 1] / np.var(x)
beta
```

```python
alpha = y.mean() - beta * x.mean()
alpha
```

```python
y_ = alpha + beta * x
```

```python
MSE = ((y - y_) ** 2).mean()
MSE
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'ro', label='sample data')
plt.plot(x, y_, lw=3.0, label='linear regression')
plt.legend();
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'ro', label='sample data')
for deg in [1, 2, 3]:
    reg = np.polyfit(x, y, deg=deg)
    y_ = np.polyval(reg, x)
    MSE = ((y - y_) ** 2).mean()
    print(f'deg={deg} | MSE={MSE}')
    plt.plot(x, np.polyval(reg, x), label=f'deg={deg}')
plt.legend();
```

```python
np.random.seed(0)
x = np.linspace(-1, 1)
y = np.random.random(len(x)) * 2 - 1
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'ro', label='sample data')
for deg in [1, 5, 9, 11, 13, 15]:
    reg = np.polyfit(x, y, deg=deg)
    y_ = np.polyval(reg, x)
    MSE = ((y - y_) ** 2).mean()
    print(f'deg={deg:2d} | MSE={MSE}')
    plt.plot(x, np.polyval(reg, x), label=f'deg={deg}')
plt.legend();
```

```python
model = Sequential()
model.add(Dense(256, activation='relu', input_dim=1))
for _ in range(3):
    model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='linear'))
model.compile(loss='mse', optimizer='rmsprop')
```

```python
model.summary()
```

```python
%%time
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'ro', label='sample data')
for _ in range(1, 8):
    model.fit(x, y, epochs=500, verbose=False)
    y_ =  model.predict(x)
    MSE = ((y - y_.flatten()) ** 2).mean()
    print(f'round={_} | MSE={MSE}')
    plt.plot(x, y_, '--', label=f'round={_}')
plt.legend();
```

### Classification

```python
f = 5
n = 10
```

```python
np.random.seed(100)
```

```python
x = np.random.randint(0, 2, (n, f))
x
```

```python
y = np.random.randint(0, 2, n)
y
```

```python
model = Sequential()
model.add(Dense(256, activation='relu', input_dim=f))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='rmsprop',
             metrics=['accuracy'])
```

```python
history = model.fit(x, y, epochs=50, verbose=False)
```

```python
y_ = np.where(model.predict(x).flatten() > 0.5, 1, 0)
y_
```

```python
y == y_
```

```python
f = 10
n = 250
```

```python
np.random.seed(100)
```

```python
x = np.random.randint(0, 2, (n, f))
x[:4]
```

```python
y = np.random.randint(0, 2, n)
y[:4]
```

```python
2 ** f
```

```python
fcols = [f'f{_}' for _ in range(f)]
fcols
```

```python
data = pd.DataFrame(x, columns=fcols)
data['l'] = y
```

```python
data.info()
```

```python
grouped = data.groupby(list(data.columns))
```

```python
freq = grouped['l'].size().unstack(fill_value=0)
```

```python
freq['sum'] = freq[0] + freq[1]
```

```python
freq.head(10)
```

```python
freq['sum'].describe().astype(int)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
```

```python
model = MLPClassifier(hidden_layer_sizes=[128, 128, 128],
                      max_iter=1000, random_state=100)
```

```python
model.fit(data[fcols], data['l'])
```

```python
accuracy_score(data['l'], model.predict(data[fcols]))
```

```python
train = data[:split]
test = data[split:]
```

```python
model.fit(train[fcols], train['l'])
```

```python
accuracy_score(train['l'], model.predict(train[fcols]))
```

```python
accuracy_score(test['l'], model.predict(test[fcols]))
```
